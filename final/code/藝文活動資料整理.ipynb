# 藝文活動-所有類別

```python
link = "https://cloud.culture.tw/frontsite/trans/SearchShowAction.do?method=doFindTypeJ&category=all"
```

```python
urllib.request.urlretrieve(link,'data.json')
```

```python
with open('data.json',"r",encoding="utf-8-sig") as file: #"r"讀檔案
    data = json.load(file)
```

```python
data[11]
```

```python
UID = [i['UID'] for i in data]
title = [i['title'] for i in data]
category = [i['category'] for i in data]

start_Time = [i['showInfo'][0]['time'] for i in data]
end_Time = [i['showInfo'][0]['endTime'] for i in data]

location = [i['showInfo'][0]['location'] for i in data]
locationName = [i['showInfo'][0]['locationName'] for i in data]

latitude = [i['showInfo'][0]['latitude'] for i in data]
longitude = [i['showInfo'][0]['longitude'] for i in data]

masterUnit = [i['masterUnit'] for i in data]
```

```python
DATA = pd.DataFrame({"辨識碼":UID,
                     "活動名稱":title,
                     "活動類型":category,
                     "開始時間":start_Time,
                     "結束時間":end_Time,
                     "活動地址":location,
                     "活動地點名稱":locationName,
                     "經度":longitude,
                     "緯度":latitude,
                     "主辦單位":masterUnit})
```

```python
DATA
```

```python
DATA.drop(index=0,inplace=True)
DATA.reset_index(drop=True,inplace=True)
DATA
```

```python
def dee(name):
    
    if len(name)==1:
        anse = name[0]
    else:
        anse = name
    
    return  anse
```

```python
DATA.主辦單位 = DATA.主辦單位.apply(dee)
```

```python
cat_dict={'1':'音樂',
          '2':'戲劇',
          '3':'舞蹈',
          '4':'親子',
          '5':'獨立音樂',
          '6':'展覽',
          '7':'講座',
          '8':'電影',
          '11':'綜藝',
          '13':'競賽',
          '14':'徵選',
          '15':'其他',
          '17':'演唱會',
          '19':'研習課程'}
```

```python
DATA.活動類型 = DATA.活動類型.map(cat_dict)
```

```python
DATA.to_csv("DATA.csv")
```

# 臺北市文化快遞資訊

```python
link = "https://cultureexpress.taipei/OpenData/Event/C000003"
urllib.request.urlretrieve(link,'data_taipei.json')
```

```python
with open('data_taipei.json',"r",encoding="utf-8-sig") as file: #"r"讀檔案
    data_taipei = json.load(file)
```

```python
data_taipei[50]
```

```python
UID = [i['ID'] for i in data_taipei]
title = [i['Caption'] for i in data_taipei]
category = [i['Category'] for i in data_taipei]

start_Time = [i['StartDate'] for i in data_taipei]
end_Time = [i['EndDate'] for i in data_taipei]

locationName = [i['Venue'] for i in data_taipei]

latitude = [i['Latitude'] for i in data_taipei]
longitude = [i['Longitude'] for i in data_taipei]

masterUnit = [i['Company'] for i in data_taipei]
```

```python
location = []
for i in data_taipei:
    if i['City']!=None:
        a = i['City']
        
        if i['Address']!=None:
            b = a + i['Address']
            location.append(b)
    
    else :
        location.append(i['City'])
```

```python
location
```

```python
DATA_taipei = pd.DataFrame({"辨識碼":UID,
                            "活動名稱":title,
                            "活動類型":category,
                            "開始時間":start_Time,
                            "結束時間":end_Time,
                            "活動地址":location,
                            "活動地點名稱":locationName,
                            "經度":longitude,
                            "緯度":latitude,
                            "主辦單位":masterUnit})
```

```python
li = list(DATA_taipei.活動地址.dropna().index)
```

```python
DATA_taipei = DATA_taipei.iloc[li ,:]
```

```python
DATA_taipei.reset_index(drop=True,inplace=True)
DATA_taipei = DATA_taipei.loc[DATA_taipei.活動地址.apply(lambda x : x.startswith("臺北市")),:]
DATA_taipei.reset_index(drop=True,inplace=True)
```

```python
DATA_taipei.辨識碼.value_counts()
```

```python
DATA_taipei = DATA_taipei.drop_duplicates()
DATA_taipei.reset_index(drop=True,inplace=True)
```

```python
DATA_taipei.to_csv("DATA_taipei_clear.csv",encoding='utf-8-sig')
```

# 臺南市政府文化局藝文活動訊息

```python
with open('data_tainan.json',"r",encoding="utf-8-sig") as file: #"r"讀檔案
    data_tainan = json.load(file)
```

```python
data_tainan[2]
```

```python
UID = [i+666 for i in range(len(data_tainan))]
title = [i['title'] for i in data_tainan]
category = [i['category'] for i in data_tainan]

start_Time = [i['act_date'] for i in data_tainan]
end_Time = [i['act_time'] for i in data_tainan]

location = [i['address'] for i in data_tainan]
locationName = [i['place'] for i in data_tainan]

latitude = [i['lat'] for i in data_tainan]
longitude = [i['lng'] for i in data_tainan]

masterUnit = [i['place_name'] for i in data_tainan]
```

```python
DATA_tainan = pd.DataFrame({"辨識碼":UID,
                            "活動名稱":title,
                            "活動類型":category,
                            "開始時間":start_Time,
                            "結束時間":end_Time,
                            "活動地址":location,
                            "活動地點名稱":locationName,
                            "經度":longitude,
                            "緯度":latitude,
                            "主辦單位":masterUnit})
```

```python
DATA_tainan.to_csv("DATA_tainan.csv")
```
